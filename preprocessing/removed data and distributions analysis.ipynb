```python
import warnings
warnings.simplefilter('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics
from scipy.stats import norm
```

```python
def plot_dist(df, class_, filename):
    dist = df['genus_name'].value_counts(normalize=True, sort=False)
    print('Max: ' + str(max(list(dist))), 'Min: ' + str(min(list(dist))), 'Avg: ' + str(statistics.mean(list(dist))))
    fig = plt.figure(figsize=(10, 6))
    ax = plt.axes()
    ax.set_facecolor('white')
    plt.bar(list(range(len(dist))), dist, width=1.5, color='gray')
    plt.xticks([])
    plt.yticks(fontsize=22, fontfamily='serif')
    plt.xlabel('')
    plt.ylabel('Frequency', fontsize=22, fontfamily='serif')
    plt.title(class_ + ' Species Distribution', fontsize=22, pad=20, fontfamily='serif')
    plt.savefig(filename+'.svg', bbox_inches='tight')
    plt.show()
```

```python
df = pd.read_csv('../kmerized_datasets/Polypodiopsida.Cleaned.k1.csv')
df = df.rename(columns={'genus name': 'genus_name'})
plot_dist(df, 'Polypodiopsida', 'polypod_dist_new')
```

```python
df = pd.read_csv('../kmerized_datasets/Pucciniomycetes.Cleaned.k1.csv')
df = df.rename(columns={'genus name': 'genus_name'})
plot_dist(df, 'Pucciniomycetes', 'pucci_dist_new')
```

```python
df = pd.read_csv('../kmerized_datasets/Rodentia.Cleaned.k1.csv')
plot_dist(df, 'Rodentia', 'rodentia_dist_new')
```

```python
df = pd.read_csv('../kmerized_datasets/Aves.Cleaned.k1.csv')
plot_dist(df, 'Aves', 'aves_dist_new')
```

```python
df = pd.read_csv('../kmerized_datasets/Chiroptera.Cleaned.k1.csv')
plot_dist(df, 'Chiroptera', 'chiroptera_dist_new')
```

```python
def plot_normal_dist(x, kind, class_, filename):
    h = sorted(np.array(x))
    fit = norm.pdf(h, np.mean(h), np.std(h))
    figure = plt.figure(figsize=(15, 9))
    plt.title('Density-Frequency Dist. of {} Species - {}'.format(kind, class_), fontsize=22, fontfamily='serif')
    plt.plot(h, fit, '-o', label='density dist.   mean={}'.format(round(np.mean(h), 2)))
    plt.hist(h, density=True, label='freq. dist.       std={}'.format(round(np.std(h), 2)))
    plt.xlabel('Species Count', fontfamily='serif', fontsize=22)
    plt.ylabel('Frequency', fontfamily='serif', fontsize=22)
    plt.xticks(fontfamily='serif', fontsize=22)
    plt.yticks(fontfamily='serif', fontsize=22)
    plt.legend(prop={'size': 22, 'family': 'serif'})
    plt.savefig(filename+'.svg', bbox_inches='tight')
    plt.show()
```

## Polypodiopsida: Species Distribution and Removed Samples Analysis

```python
polypod_rm_df = pd.read_csv('removed_data/Polypodiopsida.Removed.k1.csv')
print('total removed samples from Polypodiopsida: ', len(polypod_rm_df))
print('total removed species from Polypodiopsida: ', len(set(polypod_rm_df['genus name'])))
polypod_rm_df = pd.DataFrame(polypod_rm_df['genus name'].value_counts()[:30])
polypod_rm_df.columns = ['species count']
polypod_rm_df.insert(0, 'species name', polypod_rm_df.index, True)
polypod_rm_df.index = np.arange(len(polypod_rm_df))
polypod_rm_df.head()
```

```python
polypod_org_df = pd.DataFrame(pd.read_csv('../raw_data/Polypodiopsida.csv')['genus name'].value_counts())
polypod_org_df.columns = ['species count']
polypod_org_df.insert(0, 'species name', polypod_org_df.index, True)
polypod_org_df.index = np.arange(len(polypod_org_df))
polypod_org_df.head()
```

```python
joined_polypod = pd.DataFrame(pd.merge(polypod_rm_df, polypod_org_df, on='species name'))
joined_polypod.head()
```

```python
polypod_ratio = joined_polypod[['species count_x']].div(joined_polypod['species count_y'], axis=0).apply(lambda x: round(x*100, 2))
polypod_ratio.columns = ['removed %']
polypod_final = pd.concat([joined_polypod, polypod_ratio], axis=1).sort_values(by='removed %', ascending=False)
polypod_final.head()
```

```python
polypod_final.plot(x='species name', y=['removed %'], kind='bar', figsize=(15,9), legend=False, fontsize=22, color='white', edgecolor='black', hatch='xx')
plt.xticks(fontfamily='serif')
plt.yticks(fontfamily='serif')
plt.xlabel('Species', fontsize=22, fontfamily='serif')
plt.ylabel('Removed Samples Percentage (%)', fontsize=22, fontfamily='serif')
plt.savefig('polypod_removed_percentage.svg', bbox_inches='tight')
```

```python
polypod_org_df = pd.DataFrame(pd.read_csv('../raw_data/Polypodiopsida.csv')['genus name'].value_counts())
polypod_org_df.columns = ['species count']
polypod_org_df.insert(0, 'species name', polypod_org_df.index, True)
polypod_org_df.index = np.arange(len(polypod_org_df))
polypod_org_df.head()
```

```python
plot_normal_dist(polypod_org_df['species count'], 'Original', 'Polypodiopsida', 'org_polypod')
```

```python
polypod_proc_df = pd.DataFrame(pd.read_csv('../kmerized_datasets/Polypodiopsida.Cleaned.k1.csv')['genus name'].value_counts())
polypod_proc_df.columns = ['species count']
polypod_proc_df.insert(0, 'species name', polypod_proc_df.index, True)
polypod_proc_df.index = np.arange(len(polypod_proc_df))
polypod_proc_df.head()
```

```python
plot_normal_dist(polypod_proc_df['species count'], 'Processed', 'Polypodiopsida', 'proc_polypod')
```
