```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(3)
test_acc = [90.7, 93.00, 91.00]

fig, ax = plt.subplots(figsize=(10, 8))
ax.bar(x, test_acc, label='Testing Accuracy')
for i, v in enumerate(test_acc):
    plt.text(i-0.08, v+1, str(v))
    
plt.xticks(x, ('non-hierarchical RF', 'hierarchical Linear SVM', 'hierarchical RF'), fontsize= 13)
plt.legend(prop={'size': 11})
plt.yticks(np.arange(0, 110, 10))
plt.title('Comparison: Hierarchical vs Non-hierarchical', {'fontsize': 15})
plt.savefig('hierarchical_score_chart', bbox_inches='tight', dpi=300)
plt.show()
```
