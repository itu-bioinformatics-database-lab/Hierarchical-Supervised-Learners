```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import statistics
import json
```

```python
def plot_results(accuracies, error_df, y_error, save=False, filename='accuracies'):
    figure = plt.figure(figsize=(15, 10.5))
    plt.plot([0, 1, 2, 3, 4, 5, 6], accuracies[0][0], marker='o', linestyle=(0, (1, 1)), color='darkgrey')
    plt.plot([0, 1, 2, 3, 4, 5, 6], accuracies[0][1], marker='X', linestyle=(0, (3, 1, 1, 1, 1, 1)), color='grey')
    sns.boxplot(x='k-mer', y=y_error[0], data=error_df[0], hue='model', palette='Greys_r')
    plt.ylabel('f1 score (%)', fontfamily='serif', fontsize=22)
    plt.xlabel('k-mer length', fontfamily='serif', fontsize=22)
    plt.legend(loc='lower right', prop={'size': 22, 'family': 'serif'})
    plt.xticks(fontsize=22, fontfamily='serif')
    plt.yticks(fontsize=22, fontfamily='serif')
    if save:
        plt.savefig(filename+'.svg', bbox_inches='tight')
    plt.show()
```

```python
def get_accuracy(data):
    mean_accs = {}
    for k in sorted(data):
        for model in data[k]: 
            mean_accs.setdefault(model, [])
            mean_accs[model].append(statistics.mean(data[k][model]) * 100)
    return mean_accs
```

```python
def get_errors(data):
    error_scores = {}
    for k in data:
        for model in data[k]: 
            error_scores.setdefault(model, {})
            error_scores[model].setdefault(k, 0)
            error_scores[model][k] = data[k][model]
    return error_scores
```

```python
def prepare_df(json):
    df_train = pd.DataFrame(columns=['Training Accuracy', 'model', 'k-mer'])
    df_test = pd.DataFrame(columns=['Test Accuracy', 'model', 'k-mer'])
    for model in sorted(json):
        if 'train' in model:
            for k in sorted(json[model]):
                for score in json[model][k]:
                    df_train.loc[len(df_train)] = [score*100, model[:-6], 'k=' + k[-1]]
        else:
            for k in sorted(json[model]):
                for score in json[model][k]:
                    df_test.loc[len(df_test)] = [score*100, model[:-5], 'k=' + k[-1]]
    return df_train, df_test
```

```python
with open('../scores/hierarchical_f1_scores.json', 'r') as f1_scores:
    scores = json.load(f1_scores)
    h_scores = get_accuracy(scores)
    h_error = get_errors(scores)
```

```python
train_error, test_error = prepare_df(h_error)
```

```python
for i in train_error.index:
    if train_error.iloc[i, 1] == 'rf':
        train_error.iloc[i, 1] = 'hierarchicalRF'
    if test_error.iloc[i, 1] == 'rf':
        test_error.iloc[i, 1] = 'hierarchicalRF'
    if train_error.iloc[i, 1] == 'svm':
        train_error.iloc[i, 1] = 'hierarchicalSVM'
    if test_error.iloc[i, 1] == 'svm':
        test_error.iloc[i, 1] = 'hierarchicalSVM'
```

```python
acc_list = [[h_scores['rf test'], h_scores['svm test']],
            [h_scores['rf train'], h_scores['svm train']]]
plot_results(acc_list, [test_error, train_error], ['Test Accuracy', 'Training Accuracy'], True, 'hierarchical_f1score')
```

```python
x = np.arange(3)
test_acc = [90.8, 93.0, 91.1]

fig, ax = plt.subplots(figsize=(15, 7.5))
ax.bar(0, 90.8, color='white', edgecolor='black', hatch='\\\\', width=0.5)
ax.bar(1, 93.0, color='white', edgecolor='black', hatch='xx', width=0.5)
ax.bar(2, 91.1, color='white', edgecolor='black', hatch='//', width=0.5)
for i, v in enumerate(test_acc):
    plt.text(i-0.08, v+1, str(v), fontfamily='serif', fontweight='bold', fontsize=22)
    
plt.xticks(x, ('Non-Hierarchical RF', 'Hierarchical SVM', 'Hierarchical RF'), fontsize= 22, fontfamily='serif')
plt.ylabel('f1 score (%)', fontsize=22, fontfamily='serif')
plt.xlabel('Method Name', fontsize=22, fontfamily='serif')
plt.yticks(np.arange(0, 101, 20), fontsize=22, fontfamily='serif')
plt.savefig('hierarchical_score_chart.svg', bbox_inches='tight')
plt.show()
```
